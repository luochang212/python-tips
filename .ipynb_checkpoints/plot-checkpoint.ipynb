## 1. Matplotlib

### 1.1 条形图

```python
# 构造数据
labels = ['语文', '数学', '英语', '物理', '化学']
values = [77, 82, 65, 83, 79]
```

```python
import matplotlib.pyplot as plt
import matplotlib


# 设置字体，使汉字能够正常显示
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

# 设置字体颜色
plt.rcParams['text.color'] = 'black'

# 设置字体大小
matplotlib.rcParams['font.size'] = 15.0

fig, ax = plt.subplots()
fig.set_size_inches(6, 4)  # 设置图片的长和高

x_pos = range(len(labels))
ax.bar(x_pos, values, width=0.6, align='center')
ax.set_xticks(x_pos)
ax.set_xticklabels(labels)
ax.set_ylabel('分数')
ax.set_title('期中考成绩')
plt.show()
```

### 1.2 饼图

```python
# 构造数据
labels = ['零食', '通勤', '吃饭', '房租']
values = [0.1, 0.2, 0.3, 0.4]
```

```python
import matplotlib.pyplot as plt


# 设置字体，使汉字能够正常显示
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

# 设置字体颜色
plt.rcParams['text.color'] = 'black'

# 设置字体大小
matplotlib.rcParams['font.size'] = 15.0

# 每块饼和饼图中心的距离
explode = (0.15, 0.03, 0.03, 0.03)

fig1, ax1 = plt.subplots()
fig1.set_size_inches(7, 7)

ax1.pie(values, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
ax1.axis('equal')
plt.title('城市青年各项消费占主要消费的比例', fontsize=20)

plt.show()
```

### 1.3 折线图

#### 1.3.1 简单折线图

```python
# 构造数据
y = [57, 59, 48, 56, 59, 66, 59, 53, 65, 56, 53, 56]
x_label = ['一月', '二月', '三月', '四月', '五月', '六月', '七月', '八月', '九月', '十月', '十一月', '十二月']
```

```python
import matplotlib.pyplot as plt 
import matplotlib


# 设置字体，使汉字能够正常显示
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

# 设置字体颜色
plt.rcParams['text.color'] = 'black'

# 设置字体大小
matplotlib.rcParams['font.size'] = 15.0

fig, ax = plt.subplots()
fig.set_size_inches(6, 4)  # 设置图片的长和高

x = range(len(y))
plt.plot(x, y, label='1 line', linewidth=3, color='r', marker='o', markerfacecolor='blue', markersize=12) 
plt.xlabel('月份') 
ax.set_xticks(range(len(x_label)))
ax.set_xticklabels(x_label)
plt.xticks(rotation=45)
plt.ylabel('数量（只）') 
plt.title('烤鸭消费情况')
plt.show()
```

#### 1.3.2 复杂折线图

```python
import random

labels = ['烤鸭', '板鸭', '白斩鸭']

x_label = ['一月', '二月', '三月', '四月', '五月', '六月', '七月', '八月', '九月', '十月', '十一月', '十二月']

# 构造数据
y_1 = [random.random()*30+10 for i in range(len(x_label))]
y_2 = [random.random()*30-10 for i in range(len(x_label))]
y_3 = [random.random()*30 for i in range(len(x_label))]
```

```python
import matplotlib.pyplot as plt 
import matplotlib


# 设置字体，使汉字能够正常显示
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

# 设置字体颜色
plt.rcParams['text.color'] = 'black'

# 设置字体大小
matplotlib.rcParams['font.size'] = 15.0

fig, ax = plt.subplots()
fig.set_size_inches(6, 4)  # 设置图片的长和高

x_1, x_2, x_3 = range(len(y_1)), range(len(y_2)), range(len(y_3))
plt.plot(x_1,y_1,label=labels[0]) 
plt.plot(x_2,y_2,label=labels[1]) 
plt.plot(x_3,y_3,label=labels[2])
plt.xlabel('月份') 
ax.set_xticks(range(len(x_label)))
ax.set_xticklabels(x_label)
plt.xticks(rotation=45)
plt.ylabel('数量（只）') 
plt.title('鸭肉消费情况')
plt.legend() 
plt.show()
```

### 1.4 多图

```python
import random
import matplotlib.pyplot as plt

# 设置字体，使汉字能够正常显示
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

# 设置字体颜色
plt.rcParams['text.color'] = 'black'

fig, ax = plt.subplots()
fig.set_size_inches(13, 8)

x = [i for i in range(30)];

plt.subplot(221)
plt.plot(x, [random.random()*10 for i in range(len(x))])
plt.title('figure 1', fontsize=13)

plt.subplot(222)
plt.plot(x, [random.random()*10 for i in range(len(x))])
plt.title('figure 2', fontsize=13)

plt.subplot(223)
plt.plot(x, [random.random()*10 for i in range(len(x))])
plt.title('figure 3', fontsize=13)

plt.subplot(224)
plt.plot(x, [random.random()*10 for i in range(len(x))])
plt.title('figure 4', fontsize=13)

plt.show()
```

## 2. Seaborn

### 2.1 折线图

```python
# 构造数据
y_1 = [57, 59, 48, 56, 59, 66, 59, 53, 65, 56, 53, 56]
y_2 = [33, 37, 43, 32, 52, 47, 46, 47, 43, 51, 43, 48]
y_3 = [23, 29, 23, 38, 34, 29, 25, 29, 40, 28, 39, 33]

labels = ['烤鸭', '板鸭', '白斩鸭']

x_label = ['一月', '二月', '三月', '四月', '五月', '六月', '七月', '八月', '九月', '十月', '十一月', '十二月']
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sns.set(style="whitegrid")
sns.set(rc={'figure.figsize':(13,7)})

values = np.array([y_1, y_2, y_3])
values = values.T
data = pd.DataFrame(values, x_label, columns=labels[0:5])
plt.xticks(rotation=45)
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
sns.lineplot(data=data, palette="tab10", linewidth=2.5, sort=False)
```

## 3.  自适应的时间轴

在 matplotlib 下，给 x 轴传 datetime 对象，可使时间在 x 轴自适应地显示。即 x 轴标签既不会过于拥挤，也不会过于松散。

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
import math

# 设置字体，使汉字能够正常显示
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

# 设置字体颜色
plt.rcParams['text.color'] = 'black'

fig, ax = plt.subplots()
fig.set_size_inches(12, 5)

x = [(datetime.date.today()-i*datetime.timedelta(days=1)) for i in range(300)]
y = [10*math.sin(i)*random.random() for i in range(len(x))]

plt.plot(x,y)
plt.show()
```

## 4. 动态更新

指让图片一帧一帧地刷新。需要通过 `plt.ion()` 命令开启 interactive mode。刷新时间间隔用 `plt.pause()` 方法控制。 

此代码更适合于在原生 Python 环境下跑，不适合 Jupyter notebook。

```python
import numpy as np
from matplotlib import pyplot as plt


def main():
    plt.axis([-50,50,0,10000])
    plt.ion()
    plt.show()

    x = np.arange(-50, 51)
    for pow in range(1,5):
        y = [xi**pow for xi in x]
        plt.cla()
        plt.plot(x,y)
        plt.show()
        plt.pause(1)
        

if __name__ == '__main__':
    main()
```
