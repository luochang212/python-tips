# 联合分布

我们已经知道由联合密度函数，可以推出边际密度函数。那么如何由边际密度函数推出联合密度函数呢？

一般来讲，联合密度函数携带的信息量大于边际密度函数，因此无法由边际密度函数逆推联合密度函数。但是，如果各分量相互独立，我们有：

$$ P(x, y) = P(x)P(y) $$

也就是说，当各分量相互独立时，我们可以由边际密度函数求联合密度函数。

## 可视化

下面我们来画一个由正态分布和柏松分布构成的联合分布。

正态分布

```python
import numpy as np
import matplotlib.pyplot as plt
import math

u = 0  # 均值μ
u01 = -2
sig = math.sqrt(0.2)  # 标准差δ

x = np.linspace(u - 3 * sig, u + 3 * sig, 50)
y_sig = np.exp(-(x - u) ** 2 / (2 * sig ** 2)) / (math.sqrt(2 * math.pi) * sig)
print(x)
print("=" * 20)
print(y_sig)
plt.plot(x, y_sig, "r-", linewidth=2)
plt.grid(True)
plt.show()
```
